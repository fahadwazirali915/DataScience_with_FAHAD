## Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

---

## Dataset = "Iris" !

```python
flower = sns.load_dataset("iris")
flower
flower.head()
```

```python
sns.set_theme(style= "dark", color_codes= True)
sns.barplot(x="species", y="sepal_length", data=flower)
plt.show()
```

```python
sns.set_theme(style= "dark", color_codes= True)
sns.barplot(x="species", y="petal_length", data=flower)
plt.show()
```

### Add Titles !

```python
sns.set_theme(style= "white", color_codes= True)
sns.barplot(x="petal_length", y="sepal_width", data=flower)
plt.title("Plot of Flowers")
plt.show()
```

### Set-Styles !
- dark
- darkgrid
- white
- whitegrid
- ticks

#### **sns.set_theme(style= "___", color_codes= True)**

```python
sns.set_theme(style= "whitegrid", color_codes= True)
sns.barplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("Plot of Flowers")
plt.show()
```

### Size of Figure !

```python
sns.set_theme(style= "darkgrid", color_codes= True)
sns.barplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("Plot of Flowers")
plt.figure(figsize=(8,6))
plt.show()
```

## Another Dataset = "Titanic" !

```python
boat = sns.load_dataset("titanic")
boat
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.barplot(x="who", y="alone", hue="alive", data=boat)
plt.show()
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.barplot(x="sex", y="alone", hue="alive", data=boat, order=["female", "male"]) # Order
plt.show()
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.barplot(x="sex", y="alone", hue="who", data=boat, order=["female", "male"], color="red") # Coloe-Grading
plt.show()
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.barplot(x="sex", y="alone", hue="who", data=boat, order=["female", "male"], color="gray", ci=None) # Confidense Interval
plt.show()
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.barplot(x="sex", y="alone", hue="who", data=boat, order=["female", "male"], color="yellow", ci=None, saturation=2) # Saturation
plt.show()
```

```python
sns.barplot(x="alone", y="sex", hue="who", data=boat)

sns.set_style(style=None, rc=None)
sns.set_style("dark")

plt.show()
```
