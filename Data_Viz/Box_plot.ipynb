## Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

---

## Dataset = "Iris" !

```python
flower = sns.load_dataset("iris")
flower
flower.head()
```

```python
sns.set_theme(style= "dark", color_codes= True)
sns.boxplot(x="species",
            y="sepal_length",
            data=flower)
plt.show()
```

### Add Titles !

### Set-Styles !
- dark
- darkgrid
- white
- whitegrid
- ticks

#### **sns.set_theme(style= "___", color_codes= True)**

```python
sns.set_theme(style= "whitegrid", color_codes= True)
sns.boxplot(x="sepal_length", y="species", data=flower)
plt.title("Plot of Flowers")
plt.show()
```

### Size of Figure !

```python
sns.set_theme(style= "darkgrid", color_codes= True)
sns.boxplot(x="species", y="sepal_width", data=flower)
plt.title("Plot of Flowers")
plt.figure(figsize=(2,4))
plt.show()
```

## Another Dataset = "titanic" !

```python
boat = sns.load_dataset("titanic")
boat
```

```python
plt.figure(figsize=(10,8))

sns.set_theme(style="darkgrid", color_codes=True)
sns.boxplot(x="survived",
            y="age",
            hue="alive",
            palette="Set1",
            showmeans=True,
            meanprops={"marker":"*", "markersize":"12", "markeredgecolor":"yellow"},
            data=boat)

plt.title("Titanic-Incident", size=20, weight='bold')
plt.xlabel("How many Survived?", size=14)
plt.ylabel("What's Age?", size=13)
plt.show()
```
