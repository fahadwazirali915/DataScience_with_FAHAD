```python
# import Libraries !
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.figure as fig
import seaborn as sns
import plotly.express as px
```

```python
asia_water = pd.read_csv("south_asia_drinkingWater.csv")
asia_water.head(2)
```

```python
type(asia_water["Year"])
```

```python
plt.figure(figsize=(12,6))
sns.set_theme(style="darkgrid", color_codes=True)
sns.lineplot(x="Year",
            y="Value",
            hue="Area",
            data=asia_water,
            palette="Set1")
plt.title("Drinking Water in South Asia (2000-2020)", weight="bold", fontsize="16") 
plt.xlabel("Years", weight="bold") 
plt.ylabel("Values", weight="bold") 
plt.show()            
```
