```python
# import Libraries !
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
land = pd.read_csv("pakisatan_Land.csv")
land.head(3)
```

```python
land.describe()
```

```python
land.drop(["Area Code", "Element Code", "Element", "Domain", ], axis=1).head(3)
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.boxplot(x="Item",
            y="Year",
            hue="Flag",
            data=land)
plt.show()
```
