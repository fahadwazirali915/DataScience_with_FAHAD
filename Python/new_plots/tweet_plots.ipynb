```python
# import libraries!

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

> ## <span style="color:blue">- US-States Pregnancy-Birth-Abortion Rates </span>

```python
girls = pd.read_csv('US-State Pregnancy-Birth-Abortion Rates.csv')
girls.head()
```

```python
girls = girls[['Year', 'Events per 1,000 women', 'Metric']]
girls.head()
```

```python
girls = girls[girls['Year']>=1995]
girls.Year.min()
```

```python
# box-plot ! of men

sns.set_theme(style="darkgrid", color_codes=True)
sns.boxplot(data=girls,
                y="Year",
            palette="Set2")
plt.show()            
```

```python
# dist-plot !

sns.set_theme(style="darkgrid", color_codes=True)
sns.distplot(girls['Year'])
plt.show()            
```

```python
girls = girls[girls['Events per 1,000 women']<200]
```

```python
# box-plot ! of men

sns.set_theme(style="darkgrid", color_codes=True)
sns.boxplot(data=girls,
                y="Events per 1,000 women",
            palette="Set1")
plt.show()            
```

```python
# dist-plot !

sns.set_theme(style="darkgrid", color_codes=True)
sns.distplot(girls['Events per 1,000 women'])
plt.show()            
```

```python
fig = px.histogram(girls, x="Year", y="Events per 1,000 women",
                         color='Metric', barmode='group',
                         title="Pregnancy-Birth-Abortion Rates change in United States. (1995-2016)",
                         height=500)
fig.show()
```

---

```python
fig = px.histogram(girls,
                   x="Year",
                   y="Events per 1,000 women",
                   color="Metric",
                   title="Pregnancy-Birth-Abortion Rates change in United States. (1986-2016)",
                   height=500)
                   
fig.show()
```

---

```python
fig = px.pie(girls, values='Events per 1,000 women',
                    names='Metric', 
                    title='Pregnancy-Birth-Abortion Rates change in United States. (1986-2016)')
fig.show()
```
