```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
chilla = pd.read_csv("Chilla_data2_for_plots.csv")
chilla.head()
```

```python
sns.barplot(x="Gender",
            y="Age (years)-Float/Int", 
            hue="Are you Vaccinated?",
            palette="Set2",
            data=chilla)
plt.figure(figsize=(4,6))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
boat.head()
```

```python
plt.figure(figsize=(10,8))

sns.set_theme(style="darkgrid", color_codes=True)
sns.boxplot(x="survived",
            y="age",
            hue="alive",
            palette="Set1",
            showmeans=True,
            meanprops={"marker":"*", "markersize":"12", "markeredgecolor":"yellow"},
            data=boat)

plt.title("Titanic-Incident", size=20, weight='bold')
plt.xlabel("How many Survived?", size=14)
plt.ylabel("What's Age?", size=13)
plt.show()
```
