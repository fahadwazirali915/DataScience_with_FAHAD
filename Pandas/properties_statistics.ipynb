### Find Mean/Median/Mode !

```python
# Libraries !
import pandas as pd
import numpy as np
import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = np.array([1,5,10,10,15,20])
a
```

```python
np.mean(a)
```

```python
np.median(a)
```

```python
stats.mode(a)
```

```python
flowers = sns.load_dataset("iris")
flowers.head()
```

```python
flowers.to_csv("iris.csv")
```

```python
flowers = pd.read_csv("iris.csv")
flowers.head()
```

```python
flowers.describe()
```

```python
plt.figure(figsize=(10,8))

sns.set_theme(style="whitegrid", color_codes=True)
sns.scatterplot(x="sepal_width",
                y="sepal_length",
                hue="species",
                data=flowers,
                palette="Set1")
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.set_theme(style="whitegrid", color_codes=True)
sns.barplot(x="sepal_width",
                y="sepal_length",
                hue="species",
                data=flowers,
                palette="Set2")
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.set_theme(style="whitegrid", color_codes=True)
sns.boxplot(x="sepal_width",
                y="sepal_length",
                hue="species",
                data=flowers,
                dodge=True,
                palette="Set1")
plt.show()
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.lineplot(x="sepal_width",
            y="species",
            data=flowers)
plt.show()
```
