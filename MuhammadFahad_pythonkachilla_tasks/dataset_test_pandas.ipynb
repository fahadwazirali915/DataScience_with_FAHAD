```python
# import Libraries !
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dim = sns.load_dataset("diamonds")
dim
```

```python
dim.head()
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.boxplot(x="cut",
            y="depth",
            data=dim,
            palette="Set2")
plt.title("Diamonds View", weight="bold")
plt.show()
```

```python
sns.set_theme(style="whitegrid", color_codes=True)
sns.boxplot(x="color",
            y="price",
            hue="cut",
            data=dim,
            palette="Set1")
plt.title("Diamonds View", weight="bold")
plt.show()
```
